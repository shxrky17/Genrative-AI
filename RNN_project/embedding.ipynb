```python
import tensorflow
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[
    'the glass of milk',
    'glass of tea',
    'cup of tea',
    'cup of coffee',
    'your videos are good',
    'understand meaning of words'
]
```

```python
sent
```

```python
voc_size=1000
one_hot_repr=[one_hot(word ,voc_size)for word in sent]
one_hot_repr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
import pandas as pd
```

```python
sent_length=8
embeddded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
```

```python
embeddded_docs
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embeddded_docs)
```
